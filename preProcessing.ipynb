```python
import numpy as np 

import cv2 

from matplotlib import pyplot as plt 

#image = cv2.imread('/content/mriImage.jpg',1) 

image_bw = cv2.imread('/content/mriImage.jpg',0) 
```

```python
noiseless_image_bw = cv2.fastNlMeansDenoising(image_bw, None, 20, 7, 21) 

#noiseless_image_colored = cv2.fastNlMeansDenoisingColored(image,None,20,20,7,21)
```

```python
titles = ['Original Image (grayscale)','Image after removing the noise (grayscale)']
images = [image_bw, noiseless_image_bw]
plt.figure(figsize=(13,5))
for i in range(2):
    plt.subplot(2,2,i+1)
    plt.imshow(cv2.cvtColor(images[i],cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
norm_img = np.zeros((800,800))
final_img = cv2.normalize(noiseless_image_bw,  norm_img, 0, 255, cv2.NORM_MINMAX)
```

```python
titles = ['Original Image (grayscale)','Image after normalization']
images = [noiseless_image_bw, final_img]
plt.figure(figsize=(13,5))
for i in range(2):
    plt.subplot(2,2,i+1)
    plt.imshow(cv2.cvtColor(images[i],cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
from PIL import Image, ImageFilter
```

```python
gusImg = cv2.GaussianBlur(final_img,(5,5),0)
```

```python
titles = ['Original Image (grayscale)','Image after gusian filter']
images = [final_img, gusImg]
plt.figure(figsize=(13,5))
for i in range(2):
    plt.subplot(2,2,i+1)
    plt.imshow(cv2.cvtColor(images[i],cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```
